```python
import keyboard  # Instalé la librería antes y ahora importo este módulo.

# 1. Defino (creo) la función.
def calculadora_imc(): 
    
    """ 
    Calcula el Índice de Masa Corporal (IMC) apartir del peso y altura proporcionado por el usuario.

    Entrada: 
    - Peso (float): en kilogramos.
    - Altura (float): en metros.

    Proceso: 
    - Solicita al usuario su peso y altura. 
    - Calcua el IMC con la fórmula IMC = peso / (altura ** 2).
    - Clasifica el IMC en cuatro categorías ('Peso bajo', 'Peso normal', 'Sobrepeso', 'Obesidad').
    - Permite realizar múltiples cálculos y almacena el historial.

    Salida: 
    - Devuelve una lista[str] con el peso, altura, IMC, la categoría correspondiente y una breve descripción de la categoría. 
    - También imprime el historial de cálculos al final de la sesión.

    Errores: 
    - ValueError: si el usuario introduce un valor que no puede ser convertido a float para el peso y la altura.
    - KeyboardInterrupt: si el usuario interrumple el programa con Ctrl+C o ESC.

    """
    historial = []  # 2. Lista inicialmente vacía para luego ir añadiendo (.append) los resultados al historial de respuestas.

    while True:  # 3. Loop infinito hasta que un 'return' o 'break' lo pare.
        try:  # 4.1 Bloque try - except: intenta convertir el input del usuario a float. Si no puede, se va al 'except ValueError'.
            
            peso_input = input("Introduce tu peso en kg: ").strip()
            
            if keyboard.is_pressed('esc'):  # chequea si el usuario quiere salir.
                print("\nPrograma cerrado por el usuario. ¡Hasta luego!")
                break  # para el loop. 
            
            peso = float(peso_input)
            
            # chequea si el peso es negativo o menor que cero. 
            if peso <= 0:
                print("Formato erróneo. El peso debe ser un número mayor que cero. Intenta de nuevo.\n")
                continue  # si el valor del peso no es válido por cualquiera de las razones anteriores, te lo pide otra vez. No permite que el código avance. 

            altura_input = input("Introduce tu altura en metros: ").strip()
            
            if keyboard.is_pressed('esc'):  # chequea si el usuario quiere salir.
                print("\nPrograma cerrado por el usuario. ¡Hasta luego!")
                break  # para el loop. 
            
            altura = float(altura_input)
            
            # chequea si la altura es negativa o menor que cero.
            if altura <= 0:
                print("Formato erróneo. La altura debe ser un número mayor que cero. Intenta de nuevo.\n")
                continue  # si el valor del peso no es válido por cualquiera de las razones anteriores, te lo pide otra vez. No permite que el código avance.

            # 4. Calcula el IMC y lo redondea en una sola línea.
            imc = round(peso / (altura ** 2), 2)
            
            # 5. Clasifica del IMC.
            if imc < 18.5:
                clasificacion = "Peso bajo. Esto puede indicar que podrías estar por debajo del rango saludable."
            elif 18.5 <= imc <= 24.9:
                clasificacion = "Peso Normal. Esto sugiere que tu peso está dentro de un rango considerado saludable."
            elif 25 <= imc <= 29.9:
                clasificacion = "Sobrepeso. Esto puede indicar un riesgo potencial para la salud."
            elif imc >= 30:
                clasificacion = "Obesidad. Esto puede conllevar riesgos para la salud."
            
            resultado = f"Peso: {peso} kg, altura: {altura} m, IMC: {imc} - {clasificacion}"
            print(f"\n{resultado}")
            
            # 6. Añade el resultado al historial
            historial.append(resultado)

            # 7. Una vez completado el cálculo, le pregunta al usuario si quiere hacer otro.

            continuar = input("\n¿Quieres hacer otro cálculo de IMC? (s/n): ").strip().lower()

            if continuar not in ['s', 'si', 'sí']:
                print("\n¡Hasta luego! Gracias por usar la calculadora de IMC.")
                
                if historial: # Si el historial tiene algún elemento, se ejecuta (True). Si el historial no tiene ningún elemento, no se ejecuta (False).
                    print("\nHistorial de Cálculos:")
                    for calculo in historial:
                        print(f" - {calculo}")
                
                return historial  # 8. Finaliza toda la función completamente. Además, devuelve el historial al programa principal y me permite poder usarlo fuera de la función. 
        
        # 4.2. Aquí está el resto del bloque try - except:
        except ValueError:
            print("Formato erróneo. Por favor, introduce números válidos para peso y altura. Intenta de nuevo.\n")
        except KeyboardInterrupt: # si cierran el programa con Ctrl+C se para el bucle.
            print("\nPrograma cerrado por el usuario. ¡Hasta luego!")
            break  # termina el loop 'while True', entonces se cierra el programa.
```

```python
import pandas as pd

def matriz_conversion_temperatura(temperatura):
    """ 
    Crea una matriz de referencia para la conversión de temperaturas entre Celsius, Fahrenheit y Kelvin.
    La matriz completa se genera una vez y se puede consultar según sea necesario.

    Entrada: 
    - Temperatura (float): el valor numérico de la temperatura proporcionada por el usuario.

    Salida: 
    - Devuelve un dataframe de Pandas con todas las conversiones posibles.

    """
    # 1. Hago la matriz. Se calculan TODAS las conversiones aquí de una.
    d = {
        'Celsius': [
            temperatura,  # se asume que la temperatura proporcionada está en Celsius
            round((temperatura * 9 / 5) + 32, 2),
            round(temperatura + 273.15, 2)
        ],
        'Fahrenheit': [
            round((temperatura - 32) * 5 / 9, 2), 
            temperatura, # se asume que la temperatura proporcionada está en Fahrenheit
            round((temperatura - 32) * 5 / 9 + 273.15, 2)
        ],
        'Kelvin': [
            round(temperatura - 273.15, 2),  
            round((temperatura - 273.15) * 9 / 5 + 32, 2),
            temperatura # se asume que la temperatura proporcionada está en Kelvin
        ]
    }
    
    # 2. Lo vuelvo un dataframe
    df = pd.DataFrame(data=d, index=['Celsius', 'Fahrenheit', 'Kelvin'])
    
    # (muestro la matriz completa sólo para ver que está haciendo todo bien)
    print("\nMatriz Completa de Conversión:")
    print(df) 
    
    return df # 3. Me permite llamar este df fuera de la función. 
```

```python
def ejecutar_conversion():
    """ 
    Permite consultar conversiones específicas entre Celsius, Fahrenheit y Kelvin.

    Entrada: 
    - Unidad (str): la unidad inicial proporcionada por el usuario (Celsius, Fahrenheit, Kelvin).
    - Temperatura (float): el valor numérico de la temperatura proporcionada por el usuario.
    - Unidad_objetivo (str): la unidad a la cual se desea convertir la temperatura (Celsius, Fahrenheit, Kelvin).

    Proceso: 
    - Solicita al usuario la unidad de partida, el valor numérico de la temperatura y la unidad objetivo.
    - Valida que si la unidad proporcionada es Kelvin, el valor numérico no sea negativo.
    - Genera la matriz completa de conversiones utilizando la función 'matriz_conversion_temperatura'.
    - Consulta la matriz generada para obtener la conversión específica solicitada por el usuario.
    - Muestra el resultado de la conversión solicitada.

    Salida: 
    - Imprime un mensaje con la conversión solicitada ("X Celsius equivale a Y Kelvin.").
    - También muestra la matriz completa de conversiones para referencia.

    Errores: 
    - ValueError: si el usuario introduce un valor que no puede ser convertido a float para la temperatura.
    - KeyError: si el usuario introduce una unidad inválida (que no sea 'Celsius', 'Fahrenheit' o 'Kelvin').
    - ValueError: si el usuario intenta introducir un valor negativo para la unidad Kelvin.
    """
    while True:
        try:
            unidad = input("Escribe la unidad de partida (Celsius, Fahrenheit, Kelvin): ").strip().capitalize()
            temperatura = float(input("Introduce la temperatura que deseas convertir: "))

            # Check if the user selects Kelvin and inputs a negative value
            if unidad == "Kelvin" and temperatura < 0:
                print("Error: La temperatura en Kelvin no puede ser negativa. Intenta nuevamente.")
                continue

            unidad_objetivo = input("Escribe la unidad a la que deseas convertir (Celsius, Fahrenheit, Kelvin): ").strip().capitalize()

            # Creo una variable en la que se guardará el resultado de la función que genera la matriz completa.
            matriz = matriz_conversion_temperatura(temperatura)
        
            # Muestro sólo la conversión que pide el usuario (celda específica)
            try:
                resultado = matriz.loc[unidad, unidad_objetivo]
                print(f"\n{temperatura} {unidad} equivale a {resultado} {unidad_objetivo}.")
                break
            except KeyError:
                print("Unidad inválida. Asegúrate de que usas Celsius, Fahrenheit o Kelvin.")  # si no escriben bien la unidad. Chequea que el label coincida.
    
        except ValueError:
            print("Error: Por favor, introduce un número válido para la temperatura.")  # si no lo puede convertir a float
            continue 
```

## Exercici 3
Comptador de paraules d'un text.

Escriu una funció que donat un text, mostri les vegades que apareix cada paraula.

```python
def contador_de_palabras(text):

    """ 
    Cuenta la frecuencia de palabras en un texto dado, ignorando mayúsculas y signos de puntuación. 

    Entrada: 
    - Texto (str): cadena de texto a analizar que introduce el usuario.

    Proceso: 
    - Convierte todo el texto a minúscula.
    - Elimina signos de puntuación comunes (,.:;!?\"'()[]{})
    - Separa el texto en palabras y cuenta las ocurrencias de cada una.

    Salida: 
    - Devuelve un diccionario con palabras como claves y las frecuencias como valor. 

    Errores: 
    - - Detecta cuando se ingresa solo espacios o una cadena vacía y muestra un mensaje de error.

    """
    try: 
        text = text.strip()

        if not text: # 'if not' le pregunta a Python si la variable está vacía (si es un 'falsy'). Si está vacía, la condición es TRUE y muestra el mensaje de error. 
            raise ValueError ("No se puede analizar un texto vacío o compuesto solo de espacios. Por favor, introduce un texto válido.")
        
        # texto en minuscula
        text = text.lower()

        # quito signos de puntuación. Escribo el string entre comillas e irá uno por uno. 
        for signos in ",.:;!?\"'()[]{}":
            text = text.replace(signos, "")

        # metodo que divide por espacios
        palabras = text.split()

        # diccionario donde se irán contando las repeticiones en pares clave-valor
        contador = {}

        for cadapalabra in palabras:
            if cadapalabra in contador:
                contador[cadapalabra] += 1  # Busca el valor de esa palabra, si ya existe, le suma 1
            else:
                contador[cadapalabra] = 1  # Busca el valor de esa palabra, si no existe, empieza la cuenta en 1

        return contador
    
    except ValueError as error_message:
        print(error_message) # Imprime el mensaje del ValueError (lo escribí más arriba)
        return {} # Devuelvo un diccionario vacío como señal de que no hubo palabras válidas
    
```

```python
text = input("Escribe un texto: ") # Creo la variable. Tendrá un string con lo que el usuario escriba.
mis_palabras_contadas = contador_de_palabras(text) # Guarda el diccionario que devuelve la función.

# .items() me permite iterar por los pares clave-valor (los convierte en tuplas); sino, sólo iteraría por la clave
for palabra, cantidad in mis_palabras_contadas.items():
    if cantidad == 1:
        print(f"La palabra '{palabra}' aparece una vez.")
    else:
        print(f"La palabra '{palabra}' aparece {cantidad} veces.")




# La función sólo gestiona el procesamiento pero no el display.
# El display lo hago de forma separada.
```

## Exercici 4
Diccionari invers.

Resulta que el client té una enquesta molt antiga que s'emmagatzema en un diccionari i els resultats els necessita al revés, és a dir, intercanviats les claus i els valors. 

Els valors i claus en el diccionari original són únics; si aquest no és el cas, la funció hauria d'imprimir un missatge d'advertiment.

```python

from collections import Counter # importo el módulo 'counter' de la librería 'collections' 

def validar_e_invertir_dict(input_dict):

    """ 
    Invierte las claves y valores de un diccionario. 

    Entrada: 
    - Diccionario (dict): Un diccionario cuyas claves y valores quieran ser invertidas. 

    Proceso: 
    - Recorre el diccionario original.
    - Asigna cada valor original como la nueva clave y cada clave original como el nuevo valor. 

    Salida: 
    - Devuelve un nuevo diccionario con las claves y los valores invertidos.

    Errores: 
    - ValueError: si los valores del diccionario original no son únicos; ya que no podrían ser usadas como claves en el nuevo diccionario. 

    """

    # cuenta la ocurrencia de los valores del diccionario
    count_dict = Counter(input_dict.values())
    
    # encuentra las claves asociadas con valores repetidos 
    duplicate_keys = []
    for key, value in input_dict.items(): # .items() permite tratar el diccionario como tuplas
        if count_dict[value] > 1:
            duplicate_keys.append(key) # agrega las claves con valores repetidos para poder ver cuáles son. 


    # mensaje de error sólo si se consiguen claves con valores repetidos
    if duplicate_keys:
        return f"Error: Multiple keys for one value. See keys: {duplicate_keys}"
    
    # se crea el diccionario invertido siempre y cuando no hayan valores repetidos (usé el método 'dictionary comprehension' pero hay más formas***)
    dict_invertido = {value: key for key, value in input_dict.items()}
    
    # Finaliza la función y permite guardar el diccionario invertido fuera de esta función. 
    return dict_invertido
```
